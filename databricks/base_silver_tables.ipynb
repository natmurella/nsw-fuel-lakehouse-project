```python
catalog = "fuel"
# silver tables
silver_state_table = f"{catalog}.silver.state"
silver_brand_table = f"{catalog}.silver.brand"
silver_station_table = f"{catalog}.silver.station"
silver_fuel_table = f"{catalog}.silver.fuel"
silver_price_table = f"{catalog}.silver.price"
```

```python
# wipe and drop silver tables
spark.sql(f"DROP TABLE IF EXISTS {silver_price_table}")
spark.sql(f"DROP TABLE IF EXISTS {silver_fuel_table}")
spark.sql(f"DROP TABLE IF EXISTS {silver_station_table}")
spark.sql(f"DROP TABLE IF EXISTS {silver_brand_table}")
spark.sql(f"DROP TABLE IF EXISTS {silver_state_table}")
```

```python
# create state table if it does not exist
spark.sql(f"""
    create table if not exists {silver_state_table} (
        state_id bigint generated by default as identity,
        state_code string not null
    ) using delta
""")

# create brand table if it does not exist
spark.sql(f"""
    create table if not exists {silver_brand_table} (
        brand_id bigint generated by default as identity,
        brand_name string not null
    ) using delta
""")

# create station table if it does not exist
spark.sql(f"""
    create table if not exists {silver_station_table} (
        station_id bigint generated by default as identity,
        state_id bigint not null,
        brand_id bigint not null,
        station_code int not null,
        address string,
        name string,
        latitude double,
        longitude double
    ) using delta
""")

# create fuel table if it does not exist
spark.sql(f"""
    create table if not exists {silver_fuel_table} (
        fuel_id bigint generated by default as identity,
        fuel_code string not null
    ) using delta
""")

# create price table if it does not exist
spark.sql(f"""
    create table if not exists {silver_price_table} (
        price_id bigint generated by default as identity,
        station_id bigint not null,
        fuel_id bigint not null,
        price double,
        changed_at timestamp
    ) using delta          
""")
```
