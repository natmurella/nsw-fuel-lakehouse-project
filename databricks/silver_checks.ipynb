```python
catalog = "fuel"
# bronze tables
bronze_stations_table = f"{catalog}.bronze.nsw_fuel_stations_bronze"
bronze_prices_table = f"{catalog}.bronze.nsw_fuel_prices_bronze"
# silver tables
silver_state_table = f"{catalog}.silver.state"
silver_brand_table = f"{catalog}.silver.brand"
silver_station_table = f"{catalog}.silver.station"
silver_fuel_table = f"{catalog}.silver.fuel"
silver_price_table = f"{catalog}.silver.price"

stations_df = spark.read.table(bronze_stations_table)
silver_stations_df = spark.read.table(silver_station_table)

prices_df = spark.read.table(bronze_prices_table)
silver_prices_df = spark.read.table(silver_price_table)

# count and display the number of rows in both dataframes
print(f"Number of rows in bronze_stations_table: {stations_df.count()}")
print(f"Number of rows in silver_station_table: {silver_stations_df.count()}")

# count duplicates in silver_station_table
dupes = silver_stations_df.groupBy("station_id", "state_id", "brand_id", "station_code", "address", "name", "latitude", "longitude").count().filter("count > 1").count()
print(f"Number of duplicate rows in silver_station_table: {dupes}")

# count and display the number of rows in both dataframes
print(f"Number of rows in bronze_prices_table: {prices_df.count()}")
print(f"Number of rows in silver_price_table: {silver_prices_df.count()}")

# count duplicates in silver_price_table
dupes = silver_prices_df.groupBy("station_id", "fuel_id", "price", "changed_at").count().filter("count > 1").count()
print(f"Number of duplicate rows in silver_price_table: {dupes}")
```
